```python
#importing required librarues
import pandas as pd
import numpy as np

# Loading data
df = pd.read_csv("ICRISAT-District Level Data.csv")  
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
#checking the data types
df.info()
```

```python
#checking for missing values
pd.set_option('display.max_rows', 100)
df.isnull().sum()
```

```python
#Copy of cleaned data
df_clean = df.copy()
```

```python
#Converting Area (from 1000 ha to ha)
for col in df_clean.columns:
    if 'AREA (1000 ha)' in col:
        new_col = col.replace('AREA (1000 ha)', 'AREA (ha)')
        df_clean[new_col] = df_clean[col] * 1000
```

```python
#Converting Yield (from kg/ha to tons/ha)
for col in df_clean.columns:
    if 'YIELD (Kg per ha)' in col:
        new_col = col.replace('YIELD (Kg per ha)', 'YIELD (tons per ha)')
        df_clean[new_col] = df_clean[col] / 1000
```

```python
# Converting Production (from 1000 tons to tons)
for col in df_clean.columns:
    if 'PRODUCTION (1000 tons)' in col:
        new_col = col.replace('PRODUCTION (1000 tons)', 'PRODUCTION (tons)')
        df_clean[new_col] = df_clean[col] * 1000
```

```python
#Dropping old columns
cols_to_drop = [col for col in df_clean.columns if 
                'AREA (1000 ha)' in col or 
                'PRODUCTION (1000 tons)' in col or 
                'YIELD (Kg per ha)' in col]

df_clean.drop(columns=cols_to_drop, inplace=True)
```

```python
df_clean.head() #there are some negative values
```

```python
df_clean.isnull().sum()
```

```python
#replacing negative values as zero
df_clean[df_clean.select_dtypes(include='number').columns] = df_clean.select_dtypes(include='number').clip(lower=0)

```

```python
df_clean.head()
```

```python
#saving the cleaned data
df_clean.to_csv("agri_cleaned.csv", index=False)
```

1. Top 7 RICE PRODUCTION State Data(Bar_plot)

```python
import plotly.express as px

# Grouping total rice production by state
top_rice_states = (
    df_clean.groupby("State Name")["RICE PRODUCTION (tons)"].sum().sort_values(ascending=False).head(7).reset_index()
)

# Plotting
fig = px.bar(
    top_rice_states,
    x="State Name",
    y="RICE PRODUCTION (tons)",
    title="Top 7 Rice Producing States in India",
    labels={"RICE PRODUCTION (tons)": "Total Rice Production (tons)"},
    text="RICE PRODUCTION (tons)"
)
fig.show()
```

9.Millet Production (Last 50y)

```python
#Filtering for last 50 years
latest_years = sorted(df_clean['Year'].unique())[-50:]
millet_data = df_clean[df_clean['Year'].isin(latest_years)]

# Group by year and summing up Pearl Millet + Finger Millet production
millet_yearly = (
    millet_data.groupby('Year')[[
        'PEARL MILLET PRODUCTION (tons)',
        'FINGER MILLET PRODUCTION (tons)'
    ]].sum().reset_index()
)

#Adding a new column for Total Millet Production
millet_yearly['TOTAL MILLET PRODUCTION (tons)'] = (
    millet_yearly['PEARL MILLET PRODUCTION (tons)'] +
    millet_yearly['FINGER MILLET PRODUCTION (tons)']
)

#Plotting using line chart
fig = px.line(
    millet_yearly,
    x='Year',
    y='TOTAL MILLET PRODUCTION (tons)',
    title='Millet Production in India (Last 50 Years)',
    markers=True,
    template='plotly_white'
)

fig.update_traces(line=dict(color='purple', width=3))
fig.update_layout(xaxis_title='Year', yaxis_title='Millet Production (tons)')

fig.show()
```

12.Soybean Production by Top 5 States and Yield Efficiency

```python
soybean_stats = (
    df_clean.groupby('State Name')[['SOYABEAN PRODUCTION (tons)', 'SOYABEAN YIELD (tons per ha)']]
    .agg({'SOYABEAN PRODUCTION (tons)': 'sum', 'SOYABEAN YIELD (tons per ha)': 'mean'}).reset_index().sort_values(by='SOYABEAN PRODUCTION (tons)', ascending=False).head(5)
)
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Creating subplot with 1 row, 2 y-axes
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Bar for Soybean Production
fig.add_trace(
    go.Bar(
        x=soybean_stats['State Name'],
        y=soybean_stats['SOYABEAN PRODUCTION (tons)'],
        name='Soybean Production (tons)',
        marker_color='indigo'
    ),
    secondary_y=False
)

# Line for Yield Efficiency
fig.add_trace(
    go.Scatter(
        x=soybean_stats['State Name'],
        y=soybean_stats['SOYABEAN YIELD (tons per ha)'],
        name='Yield (tons/ha)',
        mode='lines+markers',
        marker=dict(color='orange')
    ),
    secondary_y=True
)

# Layout updates
fig.update_layout(
    title='Top 5 States by Soybean Production and Yield Efficiency',
    xaxis_title='State Name',
    yaxis_title='Production (tons)',
    legend_title='Metric',
    template='plotly_white'
)

fig.update_yaxes(title_text="Production (tons)", secondary_y=False)
fig.update_yaxes(title_text="Yield (tons/ha)", secondary_y=True)

fig.show()
```
